```python
import re
file=open("session_8.txt")
data=file.read()
print(data)


name=re.compile(r"[A-Z][a-z]+,?\s+(?:[A-Z][a-z]*\.?\s*)?[A-Z][a-z]+")
names=name.finditer(data)
for i in names :
    print (i)
   
    
print(" ")
    

num=re.compile(r"01\d{9}")
nums=num.finditer(data)
for i in nums :
    print (i)

    
print(" ")
    

email=re.compile("[a-zA-Z\d{+}]+@[a-zA-Z]+\.com")
emails=email.finditer(data)
for i in emails :
    print (i)

    
    
print(" ")

Names=name.findall(data)
Numbers=num.findall(data)
Emails=email.findall(data)
print("Names in this file are",Names)
print(" ")
print("Numbers in this file are",Numbers)
print(" ")
print("emails in this file are",Emails)

```
